```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm
from Naive_Minority_Game import Naive_Minority_Game 

game_setup = Naive_Minority_Game(memory_size = 8, n_players = 10001, n_simulations = 20000)
game_sim, players_profit = game_setup.simulate_game()
```

```python
cum_profits = np.cumsum(players_profit, axis = 0)
for row in tqdm.tqdm(cum_profits.T):
    plt.plot(np.arange(len(row)), row)
```

```python
n_sims = len(game_sim)
cum_average = np.cumsum(game_sim)/range(1, n_sims+1)
plt.plot(range(1, n_sims+1), cum_average, 'b')
plt.ylim([0,0.6])
plt.xlabel('Simulation number')
plt.ylabel('Cumulative game output')
plt.show()
```
