```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm
from Naive_Minority_Game import Naive_Minority_Game 

game_setup = Naive_Minority_Game(memory_size = 8, n_players = 10001, n_simulations = 20000)
game_sim, players_profit = game_setup.simulate_game()
```

```python
cum_profits = np.cumsum(players_profit, axis = 0)
for row in tqdm.tqdm(cum_profits.T):
    plt.plot(np.arange(len(row)), row)

plt.xlabel('Simulation number')
plt.ylabel('Player profit')
```

```python
n_sims = len(game_sim)
cum_average = np.cumsum(game_sim)/range(1, n_sims+1)
plt.plot(range(1, n_sims+1), cum_average, 'b')
plt.ylim([0,1])
plt.xlabel('Simulation number')
plt.ylabel('Cumulative game output')
plt.show()
```

```python
list_sims = []
players_profit_sim = []
for _ in tqdm.trange(30):
    game_setup = Naive_Minority_Game(memory_size = 8, n_players = 1001, n_simulations = 1000, verbose = False)
    game_sim, players_profit = game_setup.simulate_game()
    players_profit_sim.append(players_profit)
    list_sims.append(game_sim)
```

```python
for profit_sim in players_profit_sim:
    sorted_profits = np.sort(np.array(profit_sim).sum(axis=0))
    plt.plot(np.arange(len(sorted_profits)), sorted_profits)

plt.xlabel('Profit Rank (low = bad, high = good)')
plt.ylabel('Cumulative Profit')
```
